1. Git es un sistema de control de versiones (SVC por sus siglas en ingles System Version Control) de código abierto que permite a los desarrolladores registrar cambios en archivos y colaborar en proyectos de programación. 
Permite mantener un historial completo de todas las modificaciones realizadas en los archivos, lo que facilita la colaboración y la solución de problemas en un proyecto.

2. Diferencias entre Git y Github:

- Git es un sistema de control de versiones distribuido, mientras que GitHub es un servicio en línea basado en la web que proporciona alojamiento de repositorios Git en la nube.

- Git se ejecuta en la línea de comandos, mientras que GitHub ofrece una interfaz gráfica de usuario para interactuar con repositorios Git.

- Git es un software libre y gratuito, mientras que GitHub es un servicio comercial que ofrece planes gratuitos y de pago.

3. Hash en Git es una cadena única de 40 caracteres en formato hexadecimal que se utiliza para identificar de manera única cada elemento de la estructura de datos de un repositorio Git (commit, blob, tree). Un ejemplo es el hash generado a partir de un commit y que se representa en la salida del comando git log junto con información como la fecha de la confirmación, el autor, el mensaje de confirmación, entre otros; su hash podría ser 16e7e8b8a2327a1ef9d5c7b5a5b5d7b5a5b5d7b5.

4. Los elementos principales que componen la estructura de datos de Git son:

- Commits: registros de instantáneas (snapshots) del estado de los archivos en un momento determinado.
- Blobs: Unidad básica utilizada para almacenar el contenido de un único archivo, cuando el archivo es modificado se crea un nuevo blob con el contenido modificado.
- Etiquetas (tags): marcas que se utilizan para identificar versiones importantes del proyecto.
- Árbol de objetos (Trees): estructura análoga a directorios, y que incluye archivos (blobs), commits, y otros subdirectorios (trees).

5. Para crear un nuevo repositorio Git y realizar un seguimiento de los cambios en los archivos, los comandos de Git que se utilizan son:
- git init: inicializa un nuevo repositorio Git en un directorio local. (git clone se utiliza para clonar un repositorio ya existente).
- git status: Lista qué archivos se hayan en estado untracked, unstaged(modified) o staged.
- git log: Exhibe el historial completo de commits utilizando un formato predeterminado, con parámetros adicionales puede mostrar las ramas existentes y las relaciones entre ellas.
- git diff: permite visualizar las diferencias entre conjuntos de archivos del repositorio.

6. Falso: Git puede resolver automáticamente gran cantidad de conflictos de línea al fusionar ramas, pero no siempre es capaz de hacerlo por sí solo. A veces se requiere resolver de forma manual los conflictos para asegurarse de que el código fusionado sea correcto.

10. Falso: Git commit se utiliza para confirmar los cambios en el area staged del repositorio local, adjuntando un mensaje referenciando a la modificación. Para enviar estos cambios al repositorio remoto se hace uso de git push.
